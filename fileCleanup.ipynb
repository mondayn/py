```python
fullNames = []
sizes = []
allFiles = []

# traverse root directory, and list directories as dirs and files as files

curdir = "."
# curdir = "/home/nathan/py"
# curdir = "/Media/nathan/My Book"
# curdir = "/media/sf_C_drive/kp"

# print(os.walk(curdir))

for root, dirs, files in os.walk(curdir):
    path = root.split(os.sep)
#     print(os.path.abspath(root))
#     print((len(path) - 1) * '---', os.path.basename(root))
    for file in files:
#        print("{}{}".format(len(path) * '---', file))
        fullName = root + '/' + file
        size = os.stat(fullName).st_size

        fullNames.append(fullName)
        sizes.append(size)
        allFiles.append(file)
#         print("{}\t\t{}".format(size,fullName))

ds = pd.DataFrame({'full':fullNames, 'size': sizes, 'name': allFiles   })
ds.head(20)
# ds.size
```
