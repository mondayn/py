```python
#------------------start reuse ----------------------------#
import datetime

def printTimeDuration (msg):
    global dtStart, dtLast
    dtNow = datetime.datetime.now()
    cumDelta = dtNow - dtStart
    curDelta = dtNow - dtLast
    
    hours, remainder = divmod(curDelta.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    cumHours, cumRemainder = divmod(cumDelta.seconds, 3600)
    cumMinutes, cumSeconds = divmod(cumRemainder, 60)
    
    print(msg,'incr:','%s:%s:%s' % (hours, minutes, seconds),'ttl:','%s:%s:%s' % (cumHours, cumMinutes, cumSeconds))
    dtLast = dtNow
    

dtStart = datetime.datetime.now()
dtLast = dtStart
#------------------end reuse ----------------------------#


import time
time.sleep(1)

printTimeDuration('pt1') #usage

time.sleep(3)

printTimeDuration('pt3') #usage
```

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

periods=60
dates = pd.date_range('20130101', periods=periods)
dates
```

```python
df = pd.DataFrame(np.random.randn(periods,4), index=dates, columns=list('ABCD'))
df['A'].head(5)
```

```python
ax = df[['A']].plot(figsize = (12,2))
for index, row in df.iterrows():
     if row['A'] > 1:
        ax.annotate(row['A'],xy=(index,row['A']))

plt.show()
```
