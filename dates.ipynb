```python
import datetime
d2 = datetime.date.today() + datetime.timedelta(days=-30)
type(d2)
```

```python
#https://docs.python.org/3.6/library/datetime.html#strftime-strptime-behavior
print(datetime.datetime.strftime(d2, "%Y-%m-%d"))
#type(datetime.datetime.strftime(d2, "%d-%b-%Y"))
```

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
periods=60
dates = pd.date_range('20130101', periods=periods)
```

```python
df = pd.DataFrame(np.random.randn(periods,4), index=dates, columns=list('ABCD'))
df.plot.title="MyGraph"
df[['A']].plot(figsize = (12,2))
plt.show()
```
