```python
#from https://chrisalbon.com/python/pandas_with_seaborn.html
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()
df['x'] = random.sample(range(1, 100), 25)
df['y'] = random.sample(range(1, 100), 25)
df.head()
```

```python
#scatter
sns.lmplot('x', 'y', data=df, fit_reg=False)
```

```python
#density
sns.kdeplot(df.y)
```

```python
sns.distplot(df.x)
```

```python
#histogram
plt.hist(df.x, alpha=.3)
sns.rugplot(df.x);
```

```python
#box plot
sns.boxplot([df.y, df.x])
```

```python
#heatmap
sns.heatmap([df.y, df.x], annot=True, fmt="d")
```

```python
df.plot(figsize=(15,4))
```

```python
plt.plot(df.y)
```
