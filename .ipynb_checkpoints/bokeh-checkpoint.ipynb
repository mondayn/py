```python
import pandas as pd
import numpy as np

# generate larger table
dates = pd.date_range('20130101', periods=6)
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=['A', 'B', 'C', 'D'])
df = df.rename_axis('dte')
df = df.reset_index()
df.head()
```

```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool
output_notebook()
```

```python
x = df.dte
y = df.A

hover = HoverTool(tooltips=[
    ("X", "@x"),
    ("Y", "@y")
])

p = figure(width=850, height=300,tools=[hover],x_axis_type='datetime')
p.circle(x, y, size=7, color="firebrick", alpha=0.5)
p.line(x,y,color="blue")
show(p)
```

```python
from bokeh.models import ColumnDataSource

source = ColumnDataSource(df)
p = figure(plot_width=700, plot_height=300)
p.circle('dte','A',source=source,legend='A Text', size=10)
p.triangle('dte','B',color='red',source=source,legend='B Text',size=12)
p.legend.location = "top_left"
show(p)
```

```python
#small multiple / facet wrap
from bokeh.layouts import gridplot

hover = HoverTool()

options = dict(plot_width=400, plot_height=300, x_axis_type='datetime',
                              tools="pan,wheel_zoom,box_zoom,box_select,lasso_select")

# ,tools=[HoverTool()]
p1 = figure(**options)
p1.add_tools(HoverTool())
p1.circle('dte','A',source=source,legend='A Text', size=10)

p2 = figure(title='p2 title',**options)
p2.triangle('dte','B',color='red',source=source,legend='B Text',size=12)

p = gridplot([[ p1, p2]], toolbar_location="right")
show(p)
```
