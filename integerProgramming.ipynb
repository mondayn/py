```python
import cvxpy
import numpy as n
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
values = [5,5,3,3,5]
# inputs = cvxpy.Bool(len(values))
inputs = cvxpy.Variable(len(values),boolean=True)
target = 15
function = values * inputs - target
constraint = values * inputs - target >= 0
knapsack_problem = cvxpy.Problem(cvxpy.Minimize(function),[constraint])
knapsack_problem.solve()
# solver=cvxpy.GLPK_MI)
# print(inputs.value)

for i in inputs.value:
    print('{:.0f}'.format(i))
```
